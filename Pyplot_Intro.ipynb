```python
from matplotlib import pyplot as plt
```

```python
#data creation
import random
x = [i for i in range(1,101)]
y = []
for i in range(100):
  y.append(random.randint(1,101))
print(x)
print(y)
```

```python
#line plot
plt.plot(x,y)
plt.show()
```

```python
#scatter plot
plt.scatter(x,y)
plt.show()
```

```python
#customising graphs
plt.plot(x,y)
plt.title('Title')
plt.xlabel('X Gon Give It To Ya')
plt.ylabel('Y u do dis')
plt.show()
```

```python
#colours, shapes and lines
plt.plot(x,y, 'ms-')
plt.show
```

```python
#setp
lines = plt.plot(x,y)
plt.setp(lines, linestyle='--', linewidth=3, color='g')
plt.show()
```

```python
#multiple graphs
x = [i for i in range(1,101)]
y_1 = []
y_2 = []
for i in range(100):
  y_1.append(random.randint(1,101))
  y_2.append(random.randint(1,101))
```

```python
plt.plot(x,y_1,'gx',x,y_2,'rd')
plt.show()
```

```python
plt.plot(x,y_1,'g',x,y_2,'r')
plt.show()
```

```python
plt.subplot(2,1,1)
plt.title('Graph 1')
plt.plot(x,y_1,'c')
plt.subplot(2,1,2)
plt.title('Graph 2')
plt.plot(x,y_2,'k')
plt.suptitle('So many!')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.title('Graph 1')
plt.plot(x,y_1,'c')
plt.subplot(1,2,2)
plt.title('Graph 2')
plt.plot(x,y_2,'k')
plt.suptitle('So many!')
plt.show()
```

```python
#bar chart
bar_chart_data = {'one':5, 'two':10, 'three':7}
plots = plt.bar(bar_chart_data.keys(), bar_chart_data.values(), width=0.2)
plt.setp(plots, color='m')
plt.title('My First Bar Chart')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

```python
#pie chart
pie_chart_data = {'one':30, 'two':20, 'three':50}
explode = (0.3,0,0)
plt.pie(pie_chart_data.values(), labels=pie_chart_data.keys(), startangle=90, explode=explode, shadow=True)
plt.show()
```

```python
#historgram
historgram_dataset = []
for i in range(30):
  historgram_dataset.append(random.randint(1,11))
print(historgram_dataset)
```

```python
plt.hist(historgram_dataset, color='g')
plt.title('Histogram')
plt.xlabel('Categories')
plt.ylabel('Count')
plt.show()
```

```python
#cumulative
plt.hist(historgram_dataset, color='r', cumulative=True)
plt.title('Histogram')
plt.xlabel('Categories')
plt.ylabel('Count')
plt.show()
```

```python
#density
plt.hist(historgram_dataset, color='m', density=True)
plt.title('Histogram')
plt.xlabel('Categories')
plt.ylabel('Count')
plt.show()
```

```python
#3d plotting
from mpl_toolkits.mplot3d import Axes3D
axes = plt.axes(projection='3d')
axes.scatter3D(x,y_1,y_2, marker='^')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_zlabel('Z')
plt.show()
```
